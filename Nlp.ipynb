```python
# Exercise 4: Stop words
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download stopwords if you haven't already
nltk.download('punkt')
nltk.download('stopwords')

text = """
The goal of this exercise is to learn to remove stop words with NLTK. Stop words usually refers to the most common words in a language.
"""

# Tokenize into words
words = word_tokenize(text)

# Get English stopwords
stop_words = set(stopwords.words("english"))

# Remove stop words
filtered_words = [word for word in words if word.lower() not in stop_words]

print(filtered_words)
```

```python
# Exercise 5: Stemming
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

# Download punkt if not already downloaded
nltk.download('punkt')

text = """
The interviewer interviews the president in an interview
"""

# Initialize stemmer
stemmer = PorterStemmer()

# Tokenize text
tokens = word_tokenize(text)

# Stem each token
stemmed_tokens = [stemmer.stem(token) for token in tokens]

print(stemmed_tokens)
```

```python
# Exercise 6: Text preprocessing
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import string

# Download needed NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

def preprocess_text(text):
    # 1. Lowercase
    text = text.lower()
    
    # 2. Remove punctuation
    text = text.translate(str.maketrans("", "", string.punctuation))
    
    # 3. Tokenization
    tokens = word_tokenize(text)
    
    # 4. Stopword filtering
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    
    # 5. Stemming
    stemmer = PorterStemmer()
    stemmed_tokens = [stemmer.stem(word) for word in tokens]
    
    return stemmed_tokens


# Text given in the exercise
text = """
01 Edu System presents an innovative curriculum in software engineering 
and programming. With a renowned industry-leading reputation, the curriculum 
has been rigorously designed for learning skills of the digital world and 
technology industry. Taking a different approach than the classic teaching 
methods today, learning is facilitated through a collective and co-creative 
process in a professional environment.
"""

# Test
processed = preprocess_text(text)
print(processed)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import string
import nltk

# Download NLTK data
nltk.download("punkt")
nltk.download("stopwords")

# -----------------------------
# Preprocess Function
# -----------------------------
def preprocess_text(text):
    text = text.lower()
    text = text.translate(str.maketrans("", "", string.punctuation))
    tokens = word_tokenize(text)

    stop_words = set(stopwords.words("english"))
    tokens = [t for t in tokens if t not in stop_words]

    stemmer = PorterStemmer()
    tokens = [stemmer.stem(t) for t in tokens]

    return " ".join(tokens)

# -----------------------------
# Load labeled tweets
# -----------------------------
def load_labeled_tweets(filename):
    labels = []
    tweets = []

    with open(filename, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue

            label, text = line.split(",", 1)
            processed = preprocess_text(text)

            labels.append(label)
            tweets.append(processed)

    return labels, tweets

# -----------------------------
# Load file
# -----------------------------
labels, tweets = load_labeled_tweets("tweets_train.txt")

# -----------------------------
# Vectorize (Bag of Words)
# -----------------------------
vectorizer = CountVectorizer(max_features=500)
X_sparse = vectorizer.fit_transform(tweets)

print("Shape of the word count matrix:", X_sparse.shape)

# Convert sparse matrix to DataFrame
count_vectorized_df = pd.DataFrame.sparse.from_spmatrix(
    X_sparse,
    columns=vectorizer.get_feature_names_out()
)

# Add labels
count_vectorized_df["label"] = labels

print(count_vectorized_df.head())
```
